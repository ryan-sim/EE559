```python
from ucimlrepo import fetch_ucirepo 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.neighbors import DistanceMetric
from collections import Counter


# fetch dataset 
vertebral_column = fetch_ucirepo(id=212) 
  
#print(vertebral_column)
  
# data (as pandas dataframes) 
X = vertebral_column.data.original 
y = vertebral_column.data.targets 
  
X['class'] = X['class'].eq('Normal')

train_0 = X[X['class'] == True][:70]
train_1 = X[X['class']  == False][:140]
test_0 = X[X['class']  == True][70:]
test_1 = X[X['class']  == False][140:]

    
k = 208
N = 210

plt.figure()

maxk = [0,0,]

while k >= 1:

    predicted = 0
    correct_count = 0
    correct_neighbor_count = 0

    true_positive_count = 0
    true_negative_count = 0
    false_positive_count = 0
    false_negative_count = 0

    for index, row in test_0.iterrows():
        
        euclid_store = []

        for index0, row0 in train_0.iterrows():

            euclid_dist = DistanceMetric.get_metric('euclidean')
            distance = euclid_dist.pairwise([row0.iloc[:5], row.iloc[:5]])

            #[distance, row0.iloc[6]]
            euclid_store.append([distance[0][1], row0.iloc[6]])

        for index1, row1 in train_1.iterrows():

            euclid_dist = DistanceMetric.get_metric('euclidean')
            distance = euclid_dist.pairwise([row1.iloc[:5], row.iloc[:5]])

            euclid_store.append([distance[0][1], row0.iloc[6]])

        print(euclid_store)
        sorted_list = sorted(euclid_store, key=lambda x: x[0])
        occurence_count  = Counter([i[1] for i in sorted_list[:k]])
        print(occurence_count.most_common(1))
        
        actual = row.iloc[6]
        predicted = occurence_count.most_common(1)[0]
        predicted_count = occurence_count.most_common(2)[1]

        if predicted == actual:
            correct_count = correct_count + 1
            if predicted == True:
                true_positive_count = true_positive_count + 1
            else:
                true_negative_count = true_negative_count + 1
            correct_neighbor_count = correct_count + predicted_count
        else:
            if predicted == True:
                false_positive_count = true_positive_count + 1
            else:
                false_negative_count = true_negative_count + 1
            correct_neighbor_count = correct_count + N - predicted_count
    
    train_error = (N*k - correct_count) / (k * N)
    test_error = (N - correct_count) / N

    if test_error < maxk[0]:
        maxk = [test_error, k, true_positive_count, true_negative_count, false_positive_count, false_negative_count]

    plt.plot(train_error, k, 'b')
    plt.plot(train_error, k, 'r')

    #needs to be zero wheen k = 1
    #train error = proportion of wrong neighbors to right neighbors
    #test error is how many wrong answers to right answers
        

    k = k -3


plt.show()

true_positive_count = maxk[2]
true_negative_count = maxk[3]
false_positive_count = maxk[4]
false_negative_count = maxk[5]

#confusion matrix
TP = true_positive_count
print("The True Positive is ", TP)
TN = true_negative_count
print("The True Negative is ", TN)
FP = false_positive_count
print("The False Positive is ", FP)
FN = false_negative_count
print("The False Negative is ", FN)

TP_rate = TP / (TP + FN)
print("The True Positive Rate is ", TP_rate)
TN_rate = TN/ (TN + FP)
print("The True Positive Rate is ", TN_rate)
precision = TP / (TP + FP)
print("The Precision is ", precision)
f1_score = 2 * TP / (2* TP + FP + FN)
print("The F1-Score is ", f1_score)
'''
N = 10
while N <= 210:
    Choose first N/3 rows of class 0 and first N - N/3 rows of Class 1 in training set 
        Select optimal k from set starting from k =1 and incrementing by 5
    plot best test error rate
    N = N + 10

'''
```
